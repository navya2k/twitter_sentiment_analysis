```
import itertools
import re, string
from nltk import pos_tag
nltk.download('wordnet')
nltk.download('averaged_perceptron_tagger')
from nltk.stem.wordnet import WordNetLemmatizer
def remove_noise(tweet_tokens, stop_words = ()):

    cleaned_tokens = []

    for token, tag in pos_tag(tweet_tokens):
        token = re.sub("http[s]?://(?:[a-zA-Z]|[0-9]|[$-_@.&+#]|[!*\(\),]|"\
                       "(?:%[0-9a-fA-F][0-9a-fA-F]))+","", token)
        token = re.sub("(@[A-Za-z0-9_]+)","", token)
        token=re.sub("(#)","",token)
        token=re.sub("(RT)","",token)
        token=re.sub("[0-9]","",token) 
        token= "".join(u for u in token if u not in ("?", ".", ";", ":", "!",")","(","/","=","\\","-","\n","--","&","&&","||","$","!","%","+","_",">","<","~",","))
        
        token=re.sub( re.compile("["
        u"\U0001F600-\U0001F64F"  # emoticons
        u"\U0001F300-\U0001F5FF"  # symbols & pictographs
        u"\U0001F680-\U0001F6FF"  # transport & map symbols
        u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
        u"\U00002500-\U00002BEF"  # chinese char
        u"\U00002702-\U000027B0"
        u"\U00002702-\U000027B0"
        u"\U000024C2-\U0001F251"
        u"\U0001f926-\U0001f937"
        u"\U00010000-\U0010ffff"
        u"\u2640-\u2642" 
        u"\u2600-\u2B55"
        u"\u200d"
        u"\u23cf"
        u"\u23e9"
        u"\u231a"
        u"\ufe0f"  # dingbats
        u"\u3030"  # flags (iOS)
                           "]+", flags=re.UNICODE),"",token)
        if tag.startswith("NN"):
            pos = 'n'
        elif tag.startswith('VB'):
            pos = 'v'
        else:
            pos = 'a'

        lemmatizer = WordNetLemmatizer()
        token = lemmatizer.lemmatize(token, pos)

        if len(token) > 0 and token not in string.punctuation and token.lower() not in stop_words:
            cleaned_tokens.append(token.lower())
    return cleaned_tokens
```

```
def sentiment(polarity):
  if -4 <= polarity < -2: 
    label = 'Highly Negative' 
  elif -2 <= polarity < -0.5: 
    label = 'Moderately Negative' 
  elif -0.5<=polarity<0.5 : 
    label = 'Neutral' 
  elif 0.5<= polarity < 2: 
    label = 'Moderately Positive'                #tweet sentiment
  elif 2 <= polarity <= 4: 
    label = 'Highly Positive' 
  return (polarity, label)
d=[]
for i in k:
  d.append(sentiment(i))
print(d)
#d1=[]
#for i in s:
#  d1.append(sentiment(i))
#d1
```
